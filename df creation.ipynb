```python
import pandas as pd
import plotly.express as px
import requests as rq
import utils.constants as constants
import streamlit as st
```

```python
brandstof = "brandstof.csv"
with open(brandstof, encoding="utf8") as file1:
    df1 = pd.read_csv(file1, usecols=["Kenteken", "Brandstof omschrijving"])
df1.head()
```

```python
kenteken = "kenteken.csv"
with open(kenteken, encoding="utf8") as file2:
    df2 = pd.read_csv(file2, parse_dates=["Datum tenaamstelling"], usecols=["Kenteken", "Voertuigsoort", "Merk", "Handelsbenaming", "Datum tenaamstelling"])
df2.head()
```

```python
df = pd.merge(df1, df2, on='Kenteken')
df = df.dropna(subset = ["Datum tenaamstelling"])
df.head()
```

```python
df = df.set_index("Datum tenaamstelling")
```

```python
#duplicates op basis van nummerbord droppen
df = df.drop_duplicates(subset = ["Kenteken"], keep = "last")

#check
df[df.duplicated(subset = ["Kenteken"], keep = "last" )].sort_values("Kenteken")
```

```python
import pandas as pd
csv = "alltime custom.csv"
with open(csv, encoding="utf8") as file2:
    df_clean = pd.read_csv(file2, index_col="Datum tenaamstelling")
print(df_clean.shape)
df_clean.head()
```

```python
import plotly.express as px

#merken = df_clean["Merk"].unique()
#merk = st.selectbox("Merk", merken)
merk = "CITROEN"
#[df_clean["Merk"] == merk]

df_plot = df_clean[df_clean["Merk"] == merk].groupby(["Brandstof omschrijving", "Datum tenaamstelling"])["Kenteken"].count().groupby("Brandstof omschrijving").cumsum()
df_plot = df_plot.to_frame()
df_plot = df_plot.reset_index(level = [0,1]).set_index("Datum tenaamstelling").sort_index()

fuel_color_map = {'Benzine' : "brown", 'Diesel': "black", 'LPG' : "orange", 'Elektriciteit': "blue", 'CNG': "yellow", 'Alcohol' : "red",
       'Waterstof' : "aqua", 'LNG' : "green"}

fig = px.line(df_plot[:], y="Kenteken",
              color="Brandstof omschrijving",
              title="Cumulatief nieuwe registraties per brandstoftype",
              color_discrete_map = fuel_color_map)

fig.update_layout(yaxis_title="Aantal Autos cumulatief", xaxis_title = "Datum")
fig.show()

df_plot.to_csv("line " + merk +".csv")
```
